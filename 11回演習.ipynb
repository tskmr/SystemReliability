```r
library(tidyverse)
```

```r
R = function(t){
    exp(-lambda * t)
}
# lambda = 3
# R(2)
```

```r
Rs = function(t, n){
  1-(1-R(t))^n
}
lambda = 3
Rs(2, 1)
```

```r
lambda = 1

t = seq(0, 5, len = 100)
n = c(1, 2, 3)
expand.grid(t=t, n=n) %>% mutate(rel = Rs(t, n)) %>%
  ggplot(aes(x=t, y=rel, col=n, group=n)) + geom_line()
```

```r
f = function(t){
  6 * lambda * (exp(-2*lambda*t) - exp(-3*lambda*t))
}
lambda = 1
data.frame(t = seq(0, 3, len=100)) %>% mutate(dens = f(t)) %>%
  ggplot(aes(t, dens)) + geom_line() +
    geom_vline(xintercept = 5/(6*lambda), col = 2) +
      geom_vline(xintercept = log(2)/lambda, col = 3) +
              geom_vline(xintercept = (log(3)-log(2))/lambda, col = 4)
```

```r
MRL = function(t){
  1/(6*lambda) * (9-4*exp(-lambda*t))/ (3-2*exp(-lambda*t))
}
data.frame(t = seq(0, 5, len=100)) %>% mutate(mrl = MRL(t)) %>%
  ggplot(aes(t, mrl)) + geom_line()
```
